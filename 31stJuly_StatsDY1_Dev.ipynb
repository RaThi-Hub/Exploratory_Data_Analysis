# **Pearson Coefficients**

**It's the measure that helps you to determine any linear relationship, it ranges between -1 to 1**

1. If the value is between 0 - 1 --> Positive Correlation
    * if corr > 0.5 --> Strong Postive
    * if corr between 0.3 and 0.5 --> Moderate Positive
    * if corr between 0 and 0.3 --> Weak Positive

2. If the value is 0 --> No Correlation

3. If the value is between 0 to -1 --> Negative Correlation
    * if corr is in between -1 to -0.5 --> Strong Negative
    * if corr between -0.3 and -0.5 --> Moderate Negative
    * if corr between 0 and -0.3 --> Weak Negative

```python
# it
sns.heatmap(df.corr()*100, annot = True, cmap = 'Reds')
```

# **Z-score**

**Def: the number of standard Deviation a value is from the mean for a given distribution. The negative z-score indicates the values lies below the mean, and the positive value of z-score means that it lies above the mean**

**Formula: (data - mean)/(std)**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# sample data
data = np.random.normal(loc = 50, scale=10, size = 1000)
# mean and std
mean = np.mean(data)
std = np.std(data)
print(data[:10])

# calculation for z-score'
z_scor = (data - mean) / (std)
print('Mean ', mean)
print('Std: ', std)

# print some of the z_score
print('Z_scores:', z_scor[:10])

# visualize
plt.hist(z_scor)
plt.xlabel('Z-scores')
plt.ylabel('Frequency')
plt.title('Distribution of zscore')
plt.show()
```
